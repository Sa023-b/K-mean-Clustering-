```python
df = pd.read_csv('data.csv')
df.head(20)
```

```python
df[["x","y"]] = df['location'].str.split(expand = True)
df[["endX","endY"]] = df['pass_end_location'].str.split(expand = True)
df.head(20)
```

```python

df['x'] = df.loc[:,'x'] = df.x.map(lambda x: x[1:-1]).astype(float)
df['y'] = df.loc[:,'y'] = df.y.map(lambda x: x[0:-1]).astype(float)

df['endX'] = df.loc[:,'endX'] = df.endX.map(lambda x: x[1:-1]).astype(float)
df['endY'] = df.loc[:,'endY'] = df.endY.map(lambda x: x[0:-1]).astype(float)

df = df.drop(['location', 'pass_end_location'], axis =1)
```

```python
df.head(10)
```

```python
X = np.array(df[['x','y','endX','endY']])
Kmeans = KMeans(n_clusters = 10, random_state = 100)
Kmeans.fit(X)
df['cluster'] = Kmeans.predict(X)
```

```python
df.head()
```

```python
df.cluster.value_counts()
```

```python
pitch = Pitch(pitch_type='statsbomb')  # example plotting an Opta/ Stats Perform pitch
fig, ax = pitch.draw()

pitch.draw(ax=ax)

for x in range(len(df['cluster'])):
    
    if df['cluster'][x] == 0:
        pitch.lines(xstart = df['x'][x], ystart = df['y'][x], xend = df['endX'][x],yend = df['endY'][x],
                   color = '#74c69d', lw = 3, zorder =2, comet = True , ax=ax)
        
    if df['cluster'][x] == 1:
        pitch.lines(xstart = df['x'][x], ystart = df['y'][x], xend = df['endX'][x],yend = df['endY'][x],
                   color = '#38383b', lw = 3, zorder =2, comet = True , ax=ax)
        
    if df['cluster'][x] == 2:
        pitch.lines(xstart = df['x'][x], ystart = df['y'][x], xend = df['endX'][x],yend = df['endY'][x],
                   color = '#F0F8FF', lw = 3, zorder =2, comet = True , ax=ax)
        
    if df['cluster'][x] == 3:
        pitch.lines(xstart = df['x'][x], ystart = df['y'][x], xend = df['endX'][x],yend = df['endY'][x],
                   color = '#0000FF', lw = 3, zorder =2, comet = True , ax=ax)
        
    if df['cluster'][x] == 4:
        pitch.lines(xstart = df['x'][x], ystart = df['y'][x], xend = df['endX'][x],yend = df['endY'][x],
                   color = '#A52A2A', lw = 3, zorder =2, comet = True , ax=ax)
        
    if df['cluster'][x] == 5:
        pitch.lines(xstart = df['x'][x], ystart = df['y'][x], xend = df['endX'][x],yend = df['endY'][x],
                   color = '#EEC591', lw = 3, zorder =2, comet = True , ax=ax)
        
    if df['cluster'][x] == 6:
        pitch.lines(xstart = df['x'][x], ystart = df['y'][x], xend = df['endX'][x],yend = df['endY'][x],
                   color = '#53868B', lw = 3, zorder =2, comet = True , ax=ax)
        
        
    if df['cluster'][x] == 7:
        pitch.lines(xstart = df['x'][x], ystart = df['y'][x], xend = df['endX'][x],yend = df['endY'][x],
                   color = '#7FFF00', lw = 3, zorder =2, comet = True , ax=ax)
        
    if df['cluster'][x] == 8:
        pitch.lines(xstart = df['x'][x], ystart = df['y'][x], xend = df['endX'][x],yend = df['endY'][x],
                   color = '#6495ED', lw = 3, zorder =2, comet = True , ax=ax)
        
    if df['cluster'][x] == 9:
        pitch.lines(xstart = df['x'][x], ystart = df['y'][x], xend = df['endX'][x],yend = df['endY'][x],
                   color = 'black', lw = 3, zorder =2, comet = True , ax=ax)
    
```
